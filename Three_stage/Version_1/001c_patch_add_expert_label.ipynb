# 001c_patch — 補充 expert_label 欄位

不需要重新跑 HeAR，只需把 `expert_label` 從
`prepared_train_coughvid_list.csv` 合併進 `prepared_train_hear.csv`。

## Imports

```python
import pandas as pd

df_hear = pd.read_csv('data/prepared_train_hear.csv')
df_list = pd.read_csv('data/prepared_train_coughvid_list.csv')

print('df_hear columns:', df_hear.columns.tolist()[-6:])
print('df_list columns:', df_list.columns.tolist())
```

## 確認 filename 對應

```python
print('df_hear filename 範例:', df_hear['filename'].head(3).tolist())
print('df_list filename 範例:', df_list['filename'].head(3).tolist())
print('df_hear 筆數:', len(df_hear))
print('df_list 筆數:', len(df_list))
```

## 合併 expert_label

```python
# 用 filename 對應合併 expert_label
df_merged = df_hear.merge(
    df_list[['filename', 'expert_label']],
    on='filename',
    how='left'
)

print('合併後欄位（最後幾欄）:', df_merged.columns.tolist()[-8:])
print('expert_label NaN 數量:', df_merged['expert_label'].isna().sum())
print('\nexpert_label 分佈:')
print(df_merged['expert_label'].value_counts())
```

## 儲存

```python
df_merged.to_csv('data/prepared_train_hear.csv', index=False)
print('已更新 data/prepared_train_hear.csv')
print('欄位數:', len(df_merged.columns))
print('筆數:', len(df_merged))
```
